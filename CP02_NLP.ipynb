```python
import spacy
import gensim
from gensim import corpora

nlp = spacy.load("pt_core_news_sm")

with open("noticia.txt", "r", encoding="utf-8") as file:
    texto = file.read()

doc = nlp(texto)

tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct]

dictionary = corpora.Dictionary([tokens])

corpus = [dictionary.doc2bow(tokens)]

lda_model = gensim.models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)

for index, topic in lda_model.print_topics():
    print(f"Tópico {index + 1}:")
    print(topic)
    print()
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize import RegexpTokenizer

nltk.download('punkt')
nltk.download('stopwords')

with open("noticia.txt", "r", encoding="utf-8") as file:
    texto = file.read()

sentencas = sent_tokenize(texto)

tokenizer = RegexpTokenizer(r'\w+')
palavras_filtradas = [palavra.lower() for palavra in tokenizer.tokenize(texto) if palavra.lower() not in stopwords.words('portuguese')]

frequencia_palavras = FreqDist(palavras_filtradas)

pontuacao_sentencas = {}
for sentenca in sentencas:
    for palavra in word_tokenize(sentenca.lower()):
        if palavra in frequencia_palavras:
            if sentenca not in pontuacao_sentencas:
                pontuacao_sentencas[sentenca] = frequencia_palavras[palavra]
            else:
                pontuacao_sentencas[sentenca] += frequencia_palavras[palavra]

num_sentencas_resumo = int(len(sentencas) * 0.3)
sentencas_resumo = sorted(pontuacao_sentencas, key=pontuacao_sentencas.get, reverse=True)[:num_sentencas_resumo]

resumo = ' '.join(sentencas_resumo)
print(resumo)
```
